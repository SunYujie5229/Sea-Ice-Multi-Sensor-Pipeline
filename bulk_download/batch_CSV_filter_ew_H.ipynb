```python
import os
import pandas as pd

def filter_csv_in_folder(folder):
    # 遍历文件夹中的所有CSV文件
    for fname in os.listdir(folder):
        if fname.lower().endswith(".csv"):
            fpath = os.path.join(folder, fname)
            print(f"Processing {fpath} ...")
            
            # 读取CSV
            try:
                df = pd.read_csv(fpath)
            except Exception as e:
                print(f"❌ Failed to read {fpath}: {e}")
                continue
            
            # 检查必要列
            if 'sceneName' not in df.columns or 'num_matched_points' not in df.columns:
                print(f"⚠️ {fname} missing required columns, skipped.")
                continue
            
            # 筛选条件：sceneName包含 EW 和 GRDH, 且 num_matched_points > 50
            mask = (
                df['sceneName'].astype(str).str.contains('EW') &
                # df['sceneName'].astype(str).str.contains('GRDH') &
                (df['num_matched_points'] > 50)
            )
            df_filtered = df[mask]
            
            # 输出新文件名
            new_fname = os.path.splitext(fname)[0] + "_filter.csv"
            new_fpath = os.path.join(folder, new_fname)
            
            # 保存
            df_filtered.to_csv(new_fpath, index=False, encoding="utf-8-sig")
            print(f"✅ Saved filtered file: {new_fpath} ({len(df_filtered)} rows)")

# 使用示例
folder = r"F:\NWP\CS2_S1_matched\temp"   # 修改为你的CSV所在目录
filter_csv_in_folder(folder)
```
