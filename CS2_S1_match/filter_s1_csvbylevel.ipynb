```python
import os
import pandas as pd

# 设置你的主文件夹路径
root_dir = r"E:\NWP\Sentinel1\Metadata"  # 例如：r"E:\Sentinel_CSVs"
output_dir = os.path.join(root_dir, "filtered_GRD_files")  # 结果保存路径
# 用于存储所有筛选后的数据
filtered_data = []

# 遍历所有文件夹及文件
for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith(".csv"):
            file_path = os.path.join(root, file)
            try:
                df = pd.read_csv(file_path)
                if 'Processing Level' in df.columns:
                    # 筛选以 "GRD" 开头的行
                    filtered_df = df[df['Processing Level'].str.startswith('GRD', na=False)]

                    if not filtered_df.empty:
                        # 构造导出路径（保留原文件名 + 加后缀）
                        relative_path = os.path.relpath(root, root_dir)
                        save_path = os.path.join(output_dir, relative_path)
                        os.makedirs(save_path, exist_ok=True)

                        output_file = os.path.join(save_path, file.replace('.csv', '_GRD_filtered.csv'))
                        filtered_df.to_csv(output_file, index=False)
                        print(f"✅ 已保存：{output_file}")
            except Exception as e:
                print(f"❌ 读取失败：{file_path}，错误：{e}")
```
