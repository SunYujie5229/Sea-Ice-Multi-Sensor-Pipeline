```python
import asf_search as asf
import pandas as pd
import json 

# 设置搜索参数
opts = asf.ASFSearchOptions(**{
    "maxResults": 10000,
    "intersectsWith": "POLYGON ((-125.0 74.0, -75.0 74.0, -75.0 77.0, -74.73 77.51, -74.28 78.06, -73.80 78.49, -73.11 78.81, -72.25 79.11, -71.01 79.39, -70.08 79.64, -69.35 79.94, -68.64 80.3, -67.8 80.56, -67.02 80.78, -66.06 80.95, -65.13 81.13, -64.33 81.27, -63.37 81.5, -62.55 81.72, -61.86 81.93, -60.86 82.1, -60.3 82.3, -60.0 82.5, -60.0 85.0, -125.0 85.0, -125.0 74.0))",
    "start": "2022-08-01T00:00:00Z",
    "end": "2022-09-30T23:59:59Z",
    "dataset": ["SENTINEL-1"]
})
```

```python
results = asf.search(opts=opts)
```

```python

print(f"找到 {len(results)} 个产品")
# 提取你需要的字段

records = []
for r in results:
    try:
        geom = r.geometry['coordinates'][0]
        near_end = geom[0]
        near_start= geom[1]
        far_start= geom[2]
        far_end= geom[3]

        records.append({
            "Granule Name": r.properties['sceneName'],
            'platform': r.properties['platform'],
            'startTime': r.properties['startTime'],
            'stopTime': r.properties['stopTime'],
            'polarization': r.properties['polarization'],
            'url': r.properties['url'],
            "Sensor": r.properties['sensor'] ,
            "Beam Mode": r.properties['beamModeType'] ,
            "Orbit": r.properties['orbit'] ,
            "Path Number": r.properties['pathNumber'] ,
            "Processing Date": r.properties['processingDate'] ,
            "Processing Level": r.properties['processingLevel'] ,
            "Center Lat": r.properties['centerLat'] ,
            "Center Lon": r.properties['centerLon'] ,
            "Near Start Lat": near_start[1], "Near Start Lon": near_start[0],
            "Far Start Lat": far_start[1], "Far Start Lon": far_start[0],
            "Far End Lat": far_end[1], "Far End Lon": far_end[0],
            "Near End Lat": near_end[1], "Near End Lon": near_end[0],
            "Ascending or Descending": r.properties['flightDirection'] 

        })
    except Exception as e:
        print(f"⚠️ 跳过一个产品（可能缺字段）：{e}")

# 第四步：保存为CSV
df = pd.DataFrame.from_records(records)
# 保存为 CSV 文件
df.to_csv("sentinel1_ArcticCanadaNorth_2022_temp.csv", index=False)
print("✅ CSV 已保存：sentinel1_ArcticCanadaNorth_2022_temp.csv")
```
