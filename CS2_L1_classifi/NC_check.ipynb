```python
import xarray as xr

# 读取文件
file_path = r"F:\NWP\CS2_L1\2022\CS_OFFL_SIR_SIN_1B_20220103T131154_20220103T131409_E001.nc" # 替换为你的文件名
ds = xr.open_dataset(file_path)

# 方式 A: 直接打印 Dataset 对象 (在 Jupyter 中会显示交互式界面)
# 这将展示 Dimensions, Coordinates, Data variables, Attributes
display(ds) 

# 方式 B: 仅获取所有变量名的列表
# 注意：xarray 区分 "Data variables" (数据变量) 和 "Coordinates" (坐标变量)
var_names = list(ds.data_vars)
print("数据变量名:", var_names)

# 如果你也想看坐标变量 (如 lat, lon, time)
all_vars = list(ds.variables)
print("所有变量名 (含坐标):", all_vars)
```
